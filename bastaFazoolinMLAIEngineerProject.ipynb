```python
class Menu:
    def __init__(self, name, items, start_time, end_time):
        self.name = name
        self.items = items
        self.start_time = start_time
        self.end_time = end_time

    def __repr__(self):
        return f"{self.name.title()} Menu available from {self.start_time}:00 to {self.end_time}:00"

    def calculate_bill(self, purchased_items):
        total = 0
        for item in purchased_items:
            if item in self.items:
                total += self.items[item]
        return total

# Menus
menu_brunch = {
    'pancakes': 7.50, 'waffles': 9.00, 'burger': 11.00, 'home fries': 4.50, 'coffee': 1.50,
    'espresso': 3.00, 'tea': 1.00, 'mimosa': 10.50, 'orange juice': 3.50
}
menu_early_bird = {
    'salumeria plate': 8.00, 'salad and breadsticks (serves 2, no refills)': 14.00,
    'pizza with quattro formaggi': 9.00, 'duck ragu': 17.50,
    'mushroom ravioli (vegan)': 13.50, 'coffee': 1.50, 'espresso': 3.00,
}
menu_dinner = {
    'crostini with eggplant caponata': 13.00, 'caesar salad': 16.00,
    'pizza with quattro formaggi': 11.00, 'duck ragu': 19.50,
    'mushroom ravioli (vegan)': 13.50, 'coffee': 2.00, 'espresso': 3.00,
}
menu_kids = {
    'chicken nuggets': 6.50, 'fusilli with wild mushrooms': 12.00, 'apple juice': 3.00
}

# Menu instances
brunch = Menu("brunch", menu_brunch, 11, 16)
early_bird = Menu("early_bird", menu_early_bird, 15, 18)
dinner = Menu("dinner", menu_dinner, 17, 23)
kids = Menu("kids", menu_kids, 11, 21)

print(brunch)
print(brunch.calculate_bill(['pancakes', 'home fries', 'coffee']))
print(early_bird.calculate_bill(['salumeria plate', 'mushroom ravioli (vegan)']))

class Franchise:
    def __init__(self, address, menus):
        self.address = address
        self.menus = menus

    def __repr__(self):
        return f"{self.address}."

    def available_menus(self, time):
        return [menu for menu in self.menus if menu.start_time <= time <= menu.end_time]

flagship_store = Franchise("1232 West End Road", [brunch, early_bird, dinner, kids])
new_installment = Franchise("12 East Mulberry Street", [brunch, early_bird, dinner, kids])

print(flagship_store.available_menus(12))  # Noon
print(flagship_store.available_menus(17))  # 5 PM

class Business:
    def __init__(self, name, franchises):
        self.name = name
        self.franchises = franchises
    
    def get_all_menu_names(self):
        menu_names = set()
        for franchise in self.franchises:
            for menu in franchise.menus:
                menu_names.add(menu.name)
        return list(menu_names)

    def get_menu_by_name(self, menu_name):
        found_menus = []
        for franchise in self.franchises:
            for menu in franchise.menus:
                if menu.name.lower() == menu_name.lower():
                    found_menus.append(menu)
        return found_menus

    def get_menus_available_at(self, time):
        availability = {}
        for franchise in self.franchises:
            available = franchise.available_menus(time)
            availability[franchise.address] = available
        return availability


arep = Business("Basta Fazoolin' with my Heart", [flagship_store, new_installment])

menu_arepa = {
    'arepa pabellon': 7.00, 'pernil arepa': 8.50, 'guayanes arepa': 8.00, 'jamon arepa': 7.50
}

arepas_menu = Menu("arepas_menu", menu_arepa, 10, 20)
arepas_place = Franchise("189 Fitzgerald Avenue", [arepas_menu])
take_arepa = Business("Take a' Arepa", [arepas_place])





print(arep.get_all_menu_names())


brunch_menus = arep.get_menu_by_name("brunch")
for menu in brunch_menus:
    print(menu)

available_at_5pm = arep.get_menus_available_at(17)
for address, menus in available_at_5pm.items():
    print(f"At {address}, you can get:")
    for menu in menus:
        print(f" - {menu.name}")
```
